## **Data Cleaning**

Data cleaning merupakan proses menghapus atau memodifikasi data yang tidak lengkap, duplikat, tidak akurat, dan salah format. Data-data tersebut dihapus atau dimodifikasi untuk memastikan data yang sedang diolah adalah data berkualitas agar dapat menghasilkan keputusan yang lebih akurat.

1. Missing Values

Missing Values adalah sebuah kondisi dimana data hilang atau tidak terbaca.

```python
print((df.isna().sum() / len(df)) * 100)
```

Hasil dari kode di atas menunjukkan bahwa setiap kolom dalam dataset memiliki 0.0% missing values. Artinya, tidak ada data yang hilang atau tidak terbaca dalam dataset ini. Maka dari itu tidak perlu melakukan penanganan terhadap missing values (nilai yang hilang atau tidak terbaca)

2. Duplicated Values

Duplicated Values adalah sebuah kondisi dimana ada data yang muncul beberapa kali dalam satu data set.

```python
df[df.duplicated()]
```

Hasil dari kode diatas menunjukkan bahwa tidak ada yang muncul beberapa kali dalam dataset. Maka dari itu tidak perlu melakukan penanganan duplicated values dengan menghapus semua duplikasi

3. Outliers

Outliers adalah nilai yang jauh berbeda dari nilai lainnya dalam dataset. Nilai Outlier bisa jauh lebih rendah atau lebih tinggi. Outlier bisa terjadi karena berbagai alasan seperti faktor kesalahan maupun kejadian lain yang tidak terduga.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Hasil dari kode diatas menunjukkan bahwa tidak terdapat nilai yang jauh berbeda dalam dataset. Maka dari itu tidak diperlukan penanganan lebih lanjut terhadap outliers.

4. Inconsistent Values

Inconsistent Values adakah nilai-nilai yang munculnya tidak konsistem dalam dataset

6. Construct Data

Penambahan kolom baru diambil berdasarkan studi kasus yang sedang dianalisis yaitu "Analisis Faktor yang Mempengaruhi Performa Akademik Siswa Berdasarkan Gaya Belajar."

Kolom - kolom tersebut adalah:

1. Learning Efficiency Score â†’ Skor efisiensi belajar berdasarkan gaya belajar, jam belajar, dan kursus online
2. Engagement Score â†’ Skor partisipasi berdasarkan keterlibatan dalam diskusi, penggunaan teknologi pendidikan, dan penyelesaian tugas.

```python
# Menentukan skor efisiensi belajar berdasarkan gaya belajar, jam belajar, dan kursus online
def learning_efficiency(row):
    # Hitung efisiensi dasar berdasarkan gaya belajar
    if row['Preferred_Learning_Style'] == 'Visual':
        base_score = row['Study_Hours_per_Week'] * 1.2
    elif row['Preferred_Learning_Style'] == 'Auditory':
        base_score = row['Study_Hours_per_Week'] * 1.1
    elif row['Preferred_Learning_Style'] == 'Reading/Writing':
        base_score = row['Study_Hours_per_Week'] * 1.3
    elif row['Preferred_Learning_Style'] == 'Kinesthetic':
        base_score = row['Study_Hours_per_Week'] * 1.15
    else:
        base_score = row['Study_Hours_per_Week']

    # Tambahkan pengaruh jumlah kursus online yang diselesaikan
    online_course_factor = 1 + (row['Online_Courses_Completed'] * 0.02)  # Setiap kursus menambah 2%

    # Hitung skor efisiensi belajar akhir
    efficiency_score = base_score * online_course_factor
    return efficiency_score

# Terapkan ke dataset
df['Learning_Efficiency_Score'] = df.apply(learning_efficiency, axis=1)

# Menghitung skor keterlibatan siswa
def engagement_score(row):
    score = 0
    if row['Participation_in_Discussions'] == 'Yes':
        score += 2
    if row['Use_of_Educational_Tech'] == 'Yes':
        score += 2
    score += (row['Assignment_Completion_Rate (%)'] / 50)  # Normalisasi ke skala yang lebih kecil
    return score

df['Engagement_Score'] = df.apply(engagement_score, axis=1)

# Menampilkan beberapa data baru
print(df[['Study_Hours_per_Week', 'Participation_in_Discussions', 'Learning_Efficiency_Score', 'Engagement_Score']].head())
```

Berdasarkan hasil analisis yang dilakukan, terdapat beberapa temuan penting terkait faktor yang mempengaruhi performa akademik siswa berdasarkan gaya belajar:

1. Efisiensi Belajar Dipengaruhi oleh Gaya Belajar, Jam Belajar, dan Kursus Online

* Gaya belajar yang berbeda memiliki tingkat efisiensi yang bervariasi. Misalnya, siswa dengan gaya belajar Reading/Writing cenderung memiliki efisiensi lebih tinggi dibandingkan gaya lainnya karena memiliki faktor pengali tertinggi.

* Jam belajar per minggu sangat mempengaruhi skor efisiensi belajar. Siswa yang menghabiskan lebih banyak waktu belajar per minggu cenderung memiliki skor efisiensi lebih tinggi.

* Jumlah kursus online yang diselesaikan memberikan kontribusi tambahan terhadap efisiensi belajar. Setiap kursus yang diselesaikan meningkatkan efisiensi belajar sebesar 2%.



2. Keterlibatan Siswa Berkontribusi terhadap Engagement Score

* Siswa yang aktif berpartisipasi dalam diskusi dan memanfaatkan teknologi pendidikan cenderung memiliki skor keterlibatan (Engagement Score) lebih tinggi.

* Penyelesaian tugas (Assignment Completion Rate) juga menjadi faktor dalam engagement score, di mana semakin tinggi tingkat penyelesaian tugas, semakin tinggi skor keterlibatan siswa.

3. Perbandingan Efisiensi dan Keterlibatan

* Dari hasil yang diperoleh, terlihat bahwa siswa dengan jam belajar yang tinggi memiliki skor efisiensi belajar lebih tinggi, namun tidak selalu memiliki engagement score yang tinggi.

* Misalnya, siswa dengan skor efisiensi belajar tertinggi (70.656) memiliki engagement score yang juga tinggi (6.00), menunjukkan bahwa keterlibatan dalam diskusi dan penggunaan teknologi pendidikan berperan dalam meningkatkan efektivitas belajar.

* Sebaliknya, ada siswa dengan skor efisiensi belajar tinggi tetapi engagement score yang lebih rendah, yang mungkin menunjukkan bahwa mereka lebih fokus pada belajar mandiri tanpa banyak keterlibatan dalam diskusi atau teknologi pendidikan.
